```python
import pandas as pd
import numpy as np

df=pd.read_csv("C:/Users/Administrator/Documents/python/class/Jupiter notebooks/DataSets/airquality.csv")
```

```python
#1.	Fetch the observations for 9 day of June

df.query("(Month==6)&(Day==9)")
```

```python
#2.	Find Average temperature for the month of June
df.query("Month==6")['Temp'].mean()
```

```python
#3.	To which day of June has the least temperature
df1=df.query("Month==6")['Temp'].min()
df3=df.query("Month==6")
df2=df3.Temp
df3[["month_val","Day","Temp"]][np.in1d(df2,df1)]
```

```python
#4.	Find Maximum Ozone value for the month of May

df.query("Month==5")['Ozone'].max()
```

```python
#5.	Find the count of the missing values in the ozone column of the data set
df['Ozone'].isnull().sum()
```

```python
#6.	Find out What is the mean of the Ozone column in this dataset

df['Ozone'].mean()
```

```python
#7.	Find out which month has the highest temperature

val=df.groupby('Month')['Temp'].max().max()
df1=df.Temp
df[["month_val","Temp"]][np.in1d(df1,val)]
```

```python
#8.	Find out the wind value when the Ozone becomes maximum

val=df['Ozone'].max()
df[['Wind','Ozone']][np.in1d(df.Ozone,val)]
```

```python
#9.	Find out the months for which the airquality observations have been carried out
import calendar

df["month_val"]=df["Month"].apply(lambda x: calendar.month_abbr[x])
df["month_val"].unique().tolist()
```

```python
#10.	Find the Ozone and temperature values for the 1st observation of every month.
df.query("Day==1")[["month_val","Day","Ozone","Temp"]]
```

```python
#11.	Which day of which month corresponds to the least Ozone Value

val=df["Ozone"].min()
df[["month_val","Day","Ozone"]][np.in1d(df.Ozone,val)]
```

```python
#12.	Convert the temperature for all the observations to Centigrade scale
df["Temp_cent"]=df["Temp"].apply(lambda x: (x-32) * 5/9)
df[["Temp","Temp_cent"]]
```
